```python
from qutip import Bloch
import numpy as np
from math import sin,cos

#defing angles
theta=np.radians(135)
phi=np.radians(0)

#Making sphere whith the given values of theta and phi
qbit=Bloch()
x=sin(theta)*cos(phi)
y=sin(theta)*sin(phi)
z=cos(theta)
qbit.add_vectors([x,y,z])

#Generating image
qbit.show()


```

```python
from qiskit import QuantumCircuit
import numpy as np


#creating 2 qbits
qbit=QuantumCircuit(2)
#applying hadamard gate for 1st qbit
qbit.h(0)
#applying NOT gate for 2nd qbit
qbit.x(1)
#applying hadamard gate for 2nd qbit
qbit.h(1)
#storing the qbits in the form of classical bits
qbit.measure_all()
print(qbit)
```
